```python
def f(x1,x2):
    return (1.5-x1+x1*x2)**2+(2.25-x1+x1*x2**2)**2+(2.625-x1+x1*x2**3)**2
```

```python
def RealF(x1):
    
    value = f(x1[0],x1[1])
    return 1/(value+1)
    
```

```python
def num_bits(bounds):
    return math.ceil(math.log((bounds[1]-(bounds[0]))*10**3,2))
```

```python
def initialPopu(bits,popu,variables):
    Pupulation =[]
    for i in range(popu):
        individual = ""
        for i in range(bits*variables):
            if np.random.uniform(0,1,1)[0] < 0.5:
                #individual.append(0)
                
                individual = individual + "0"
            else:
                #individual.append(1)
                individual = individual + "1"
        Pupulation.append(individual)
    return Pupulation
            
```

```python
initial = initialPopu(14,100,2)
initial
```

```python
def getValues(chromosomes,num_bits, variables, bounds):
    c=0
    curr = ""
    valuesR = []
    valuesN = []
    for chromosome in range(len(chromosomes)):
        for i in chromosomes[chromosome]:
            curr = curr+i
            c+=1
            if c == num_bits:
                x = int(curr,base=2)
                valuesN.append(x)
                newx = bounds[0] + ((9*x)/((2**num_bits)-1))
                valuesR.append(newx)
                curr = ""
                c = 0
        chromosomes[chromosome] = [chromosomes[chromosome],valuesN,valuesR,f(valuesR[0],valuesR[1]),RealF(valuesR)]
        valuesR = []
        valuesN = []
    
    return chromosomes
```

```python
def getOneValues(chromosomes,num_bits, variables, bounds):
    c=0
    curr = ""
    valuesR = []
    valuesN = []

    for i in chromosomes[0]:
        curr = curr+i
        c+=1
        if c == num_bits:
            x = int(curr,base=2)
            valuesN.append(x)
            newx = bounds[0] + ((9*x)/((2**num_bits)-1))
            valuesR.append(newx)
            curr = ""
            c = 0
    print(valuesR)
    new_chromosome = [chromosomes[0],valuesN,valuesR,f(valuesR[0],valuesR[1]),RealF(valuesR)]
    valuesR = []
    valuesN = []
    
    return new_chromosome
```

```python
def getMean(chromosomes,mu):
    mean = 0
    for i in chromosomes:
        mean += i[-1]
    return mean/mu
    
```

```python
def parentsSelection(chromosomes,mu):
    r = random.random()
    mean = getMean(chromosomes,mu)
    cont = 0
    parents = []
    for i in range(mu):
        cont += (chromosomes[i][-1] / mean )
        while cont > r:
            parents.append(chromosomes[i])
            r+=1
    return parents
```

```python
def combination(chro1,chro2, prob):
    r = np.random.uniform(0,1,1)[0]
    if prob > r :
        
    
        ind1 = random.randint(1,len(chro1[0]))
        ind2 = random.randint(1,len(chro1[0]))

        if ind1 == ind2:
            while True:
                ind2 = random.randint(1,len(chro1[0]))
                if ind1 != ind2:
                    break
        if ind2 < ind1: 
            ind2 = ind1
            ind1 = ind2

        new1 = "" 
        new2 = ""
        new1 = new1 + chro1[0][:ind1]
        new2 = new2 + chro2[0][:ind1]

        new1 = new1 + chro2[0][ind1:ind2]
        new2 = new2 + chro1[0][ind1:ind2]

        new1 = new1 + chro1[0][ind2:]
        new2 = new2 + chro2[0][ind2:]
        new1 = getValues([new1], 14, 2, [-4.5, 4.5])
        new2 = getValues([new2], 14, 2, [-4.5, 4.5])

        return new1,new2
    
    else: 
        return chro1,chro2,
        
    
```

```python
def mutation(arr,pm):
    chrom = arr[0]
    newchrom = ""
    for i in chrom:
        if np.random.uniform(0,1,1)[0] < pm:
            if i == '0':
                newchrom += '1'
            else:
                newchrom += '0'
        else:
            newchrom += str(i)
    return getValues([newchrom], 14, 2, [-4.5, 4.5])
```

```python
def beestworst(arr):
    best = arr[0][-1]
    best_complete_array = arr[0]
    worst =arr[0][-1]
    worst_complete_array = arr[0]
    for i in arr:
        if i[-1] > best:
            best = i[-1]
            best_complete_array = i
            
        if i[-1] < worst:
            worst = i[-1]
            worst_complete_array = i
    return best_complete_array,worst_complete_array
```

```python
def survivorseries(parents, son):
    bestB, worstB =  beestworst(parents)
    bestW, worstW =  beestworst(son)
    if bestB[-1] > worstW[-1]:
        son.remove(worstW)
        son.append(bestB)
    return son
```

```python
def AG(G,mu,pc,pm,bounds,variables):
    bits = num_bits(bounds)
    initial = initialPopu(bits,mu,variables)
    initial = getValues(initial,bits, variables, bounds)
    t = 0
    while t<=G:
        parents = parentsSelection(initial,mu)
        sons = []
        
        for i in range(0,mu,2):
            if random.random() < pc:
                h1,h2=  combination(parents[i],parents[i+1],0.9)
            else:
                h1 = parents[i].copy()
                h2 = parents[i+1].copy()
            if len(h1[0][0]) == 1:
                h1 = [h1]
                
            if len(h2[0][0]) == 1:
                h2 = [h2]
            h1 = mutation(h1[0],pm)
            h2 = mutation(h2[0],pm)
            sons.append(h1[0])
            sons.append(h2[0])
        
        initial = survivorseries(parents, sons)
        t +=1
        
    best,_ = beestworst(initial)  
    return best
    
```

```python
def combinationT(chro1,chro2,variables,bounds):
    chro1,chro2 = chro1[0],chro2[0]
    
    alpha = random.uniform(0,1)
    son1 = alpha * np.array(chro1) + (1-alpha) * np.array(chro2)
    alpha = random.uniform(0,1)
    son2 = alpha * np.array(chro2) + (1-alpha) * np.array(chro1)
    
    son1 = getValuesT([son1], variables, bounds)
    son2 = getValuesT([son2], variables, bounds)
    return son1[0],son2[0]
```

```python
def mutationT(arr,pm):
    chrom = arr[0]
    newchrom = []
    for i in chrom:
        if np.random.uniform(0,1,1)[0] < pm:
            newchrom.append(np.random.uniform(-4.5,4.5))
        else:
            newchrom.append(i)
    return getValuesT([newchrom],2, [-4.5, 4.5])
```

```python
def initialPopuTravel(popu,variables,dic):
    Pupulation =[]
    for i in range(popu):
        individual = [0]
        cities = list(range(1, variables))
        for i in range(variables):
            if len(cities) == 0:
                break
            target = random.choice(cities)
            individual.append(target)
            cities.remove(target)
        individual.append(0)##
        total = gettotal(individual,dic.copy())
        Pupulation.append([individual,total])
    return Pupulation
            
```

```python
def parentsSelectionTravel(chromosomes,mu):
    r = random.random()
    #print("chr",chromosomes)
    mean = getMeanTravel(chromosomes,mu)
    cont = 0
    parents = []
    for i in range(mu):
        cont += (chromosomes[i][-1] / mean )
        while cont > r:
            parents.append(chromosomes[i])
            r+=1
    #print("parents",parents)
    return parents
```

```python
def combinationTravel(chro1,chro2,dic):
    chro1,chro2 = chro1[0],chro2[0]
    stop1 = random.randint(1,len(chro1)-1)
    stop2 = random.randint(1,len(chro1)-1)
    lenchr = len(chro1)-1
    
    if stop1 > stop2:
        stop2df = stop2
        stop2 = stop1
        stop1 = stop2df
    
    

    son1 = [0]*lenchr
    son2 = [0]*lenchr
    son1[stop1:stop2] = chro2[stop1:stop2]
    son2[stop1:stop2] = chro1[stop1:stop2]

    
    son1range = list(range(lenchr))
    son2range = list(range(lenchr))
    
    for i in son1:
        if i in son1range:
            son1range.remove(i)
            
    for i in son2:
        if i in son2range:
            son2range.remove(i)
            
            
            


    
    
    
    for i in range(1,lenchr):
        if son1[i] == 0:
            son1[i] = son1range[0]
            son1range.pop(0)
            
    for i in range(1,lenchr):
        if son2[i] == 0:
            son2[i] = son2range[0]
            son2range.pop(0)
            

    son1.append(0)
    son2.append(0)
#    res = []
#    for i in son1:
#        if i not in res:
#            res.append(i)
#            
#            
#    res2 = []
#    for i in son2:
#        if i not in res2:
#            res2.append(i)

            

    son1 = [son1,gettotal(son1,dic.copy())]
    son2 = [son2,gettotal(son2,dic.copy())]

    return son1,son2
```

```python
def MutationTravel(son1, variables,dic):
    son1 = son1[0]
    times = random.randint(1,variables-1)
    for i in range(times):
        one = random.randint(1,variables-1)
        onevalue = son1[one]
        two = random.randint(1,variables-1)
        twovalue = son1[two]
        son1[one] = twovalue
        son1[two] = onevalue
        
    son1 = [son1,gettotal(son1,dic.copy())]
    return son1
        
    
        
```
